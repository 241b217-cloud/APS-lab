# Shortest Path Using Dijkstra Algorithm in Python
Problem Description:
Given a weighted graph with non negative edge weights. Find the shortest distance from a source vertex to all other vertices using Dijkstra algorithm

#Theory
Dijkstra algorithm finds shortest distance from one source node to all other nodes in a weighted graph with non negative weights. Dictionary structure stores nodes as keys and neighbor nodes with weights as values.

##Algorithm Steps

1. Represent graph using dictionary.
2. Set distance of every node as infinity.
3. Set distance of source node as 0.
4. Select node with smallest distance among unvisited nodes.
5. Update distance of neighbor nodes.
6. Repeat until all nodes get visited.

```python
def dijkstra(graph,start):
    dist={i:1e9 for i in graph}
    prev={i:None for i in graph}
    visited=[]
    dist[start]=0
    while len(visited)<len(graph):
        min_node=None
        min_dist=1e9
        for node in graph:
            if node not in visited and dist[node]<min_dist:
                min_dist=dist[node]
                min_node=node
        if min_node==None:
            break
        visited.append(min_node)
        for nei,wt in graph[min_node].items():
            if dist[min_node]+wt<dist[nei]:
                dist[nei]=dist[min_node]+wt
                prev[nei]=min_node
    return dist,prev
def getpath(prev,start,end):
    path=[]
    while end!=None:
        path.append(end)
        end=prev[end]
    path=path[::-1]
    if path[0]==start:
        return path
    return []
graph={}
n=int(input("Enter number of nodes: "))
for _ in range(n):
    node=input("\nEnter node name: ")
    graph[node]={}
    k=int(input(f"Enter number of neighbors for {node}: "))
    for _ in range(k):
        nei=input(f"Enter neighbor of {node}: ")
        wt=int(input(f"Enter weight from {node} to {nei}: "))
        graph[node][nei]=wt
print("\n--- Enter 2 additional nodes ---")
for _ in range(2):
    node=input("\nEnter node name: ")
    graph[node]={}
    k=int(input(f"Enter number of neighbors for {node}: "))
    for _ in range(k):
        nei=input(f"Enter neighbor of {node}: ")
        wt=int(input(f"Enter weight from {node} to {nei}: "))
        graph[node][nei]=wt
start=input("\nEnter starting node: ")
dist,prev=dijkstra(graph,start)
print("\n--- Shortest Paths ---")
for node in graph:
    p=getpath(prev,start,node)
    if dist[node]==1e9:
        print(start,"to",node,"no path")
    else:
        print(start,"to",node,"=",int(dist[node]),"path","->".join(p))
```

```python
graph = {
'A': {'B': 4, 'C': 2},
'B': {'A': 4, 'C': 5, 'D': 10},
'C': {'A': 2, 'B': 5, 'D': 3},
'D': {'B': 10, 'C': 3}
}

result = dijkstra(graph, 'A')

for node in result:
    print("Distance from A to", node, "=", result[node])
```
